Import bibliotek

```python
import numpy as np
import pandas as pd
import sklearn
sklearn.__version__
```

Generowanie danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'grren', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

Utworzenie kopii danych

```python
df = df_raw.copy()
df.info()
```

Zmiana typu danych i wstępna eksploracja

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

```python
df.describe(include=['category']).T
```

Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
le.classes_
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
le.inverse_transform(df['bought'])
df['bought'] = le.inverse_transform(df['bought'])
```

OneHotencoder

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

Pandas get_dummies

```python
pd.get_dummies(data=df, drop_first=True, dtype=int)
```

Standaryzacja

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie strandardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
from sklearn.preprocessing import StandardScaler

scaler_price = StandardScaler()
scaler_price.fit(df[['price']])
df['price'] = scaler_price.transform(df[['price']])
```

```python
scaler_weight = StandardScaler()
df['weight'] = scaler_weight.fit_transform(df[['weight']])
```

```python
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
```
